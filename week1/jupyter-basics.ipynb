## 주피터 노트북 사용법
* Shift + Enter : 셀이 실행되고 커서가 다음 셀로 이동합니다.
* Enter : 다시 편집 상태로 돌아옵니다.
* ESC를 누르고
    * a : 위에 셀이 추가됩니다.
    * b : 아래에 셀이 추가됩니다.
    * dd : 셀이 삭제됩니다.
    * m : Markdown 셀로 변경됩니다.
    * y : Code 셀로 변경됩니다.
* h : 단축키 모음
